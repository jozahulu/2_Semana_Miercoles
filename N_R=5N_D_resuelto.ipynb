```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=1000  # Numero de datos
rand_x= []    # Vectores de datos aleatorios
rand_y=[]
rand_x5= []    # Vectores de datos aleatorios pero por cinco
rand_y5=[]
for i in range (N):
    ale_x=np.random.uniform(0.,10.,N) #LOs datos son iguales a los de radio con la longitud de la caja
    ale_y=np.random.uniform(0.,10.,N)
    ale_x5=np.random.uniform(0.,10.,5*N)
    ale_y5=np.random.uniform(0.,10.,5*N)  #LOs datos son cinco veces mas que del a caja
    rand_x.append(ale_x)
    rand_y.append(ale_y)
    rand_x5.append(ale_x5)
    rand_y5.append(ale_y5)
```

```python
plt.title('union de todos los puntos')
plt.plot(rand_x[0],rand_y[0],'.')
plt.plot(rand_x5[1],rand_y5[1],'.')
```

```python
circle_r = 9
circle_x = 5
circle_y = 5

phi = 2 * np.pi * np.random.rand(N)
r = circle_r * np.random.rand(N)

x = np.sqrt(r) * np.cos(phi) + circle_x
y = np.sqrt(r) * np.sin(phi) + circle_y

plt.title('Grafica de  Disco')
plt.plot(x, y, '.')
plt.xlabel(r"x", fontsize = 15)
plt.ylabel(r"y", fontsize = 15)
```

```python
plt.title('Union de discos y cuadrados ')
plt.plot(rand_x[0],rand_y[0],'.')
plt.plot(rand_x5[1],rand_y5[1],'.')
plt.plot(x,y,'.',color='k')
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
#Definicion de Vectores 
dis_datos = []
dis_aleatorios = []
dis_datos5 = []
dis_aleatorios5 = []
uno = rand_x[0]
dos = rand_y[0]
tres = rand_x5[1]
cuatro = rand_y5[1]

for i in range(N):
    for j in range(i+1,N):
        d_a1 = distancia(uno[i],uno[j],dos[i],dos[j])
        d_a2 = distancia(tres[i],tres[j],cuatro[i],cuatro[j])
        
        dis_aleatorios.append(d_a1)
        dis_aleatorios5.append(d_a2)
```

```python
nbins = np.linspace(0.,np.sqrt(2)*10, 20)
plt.title('R=N el numero de datos al numero del disco')
ran_n, ran_bin,m = plt.hist(dis_aleatorios,nbins, label='$RR(r)$')
print(nbins)
plt.legend()
```
