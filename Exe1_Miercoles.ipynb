```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=2000  # Numero de Datos aleatorios 
p_x=np.random.uniform(0.,10.,N)
p_y=np.random.uniform(0.,10.,N)
plt.plot(p_x,p_y,'.',color='red')
#print(ale_x)
#print(len(ale_x))
```

```python
c_r = 25   # Radio del circulo para la variacion de puntos
c_x = 5    # Punto central en el eje x
c_y = 5    # Punto central en el eje y 
angulo = 2 * np.pi * np.random.rand(N)
r = c_r * np.random.rand(N)
```

```python
x = np.sqrt(r) * np.cos(angulo) + c_x    
y = np.sqrt(r) * np.sin(angulo) + c_y
```

```python
plt.plot(x, y, '.',color='green')
plt.xlabel(r"x", fontsize = 15)
plt.ylabel(r"y", fontsize = 15)
#plt.xlim(-1.5, 1.5)
#plt.ylim(-1.5, 1.5)
```

```python
plt.plot(p_x,p_y,'.',color='red')
plt.plot(x,y,'.',color='green')
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dtot = N*(N-1)/2
j = 0
dis_datos = []
dis_aleatorios = []
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(x[i],x[j],y[i],y[j])
        d_a = distancia(p_x[i],p_x[j],p_y[i],p_y[j])
        dis_datos.append(d_d)
        dis_aleatorios.append(d_a)
```

```python
data_n, data_bin,m = plt.hist(dis_datos,bins=20,color='red')
```

```python

ran_n, ran_bin, p = plt.hist(dis_aleatorios,bins=20,color='green')
```

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1, label=('$\dfrac{DD(r)}{RR(r)}-1$'),color='red')
plt.legend()
```

```python
dit_da = []
```

```python
for i in range(N):
    for j in range(N):
        d_dr = distancia(x[i],p_x[j],y[i],p_y[j])
        dit_da.append(d_dr)
```

```python
da_n, da_bin, q = plt.hist(dit_da,bins=20,color='blue')
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(da_n)**2,label=('$\dfrac{DD(r)RR(r)}{DR(r)^2}$'))
plt.legend()
```

```python

plt.plot(data_bin[1:],1.+ data_n/ran_n -2*da_n/ran_n, label=('Landy-Szalay=$1+\dfrac{DD(r)}{RR(r)}-2\dfrac{DR(r)}{RR(r)}$'))
```

```python
plt.plot(data_bin[1:],((data_n)/(da_n))-1,label=('Davis-Peebles=$\dfrac{DD(r)}{DR(r)}-1$'))

plt.legend()
```
